# Jupyter Notebook Example

This is an example of a Jupyter Notebook. With it, you can easily mix Markdown text, Latex math and executable python. 

For instance, in this notebook, I'll import some math related python libraries and plot them using matplotlib (a Python plotting library)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 6*2*np.pi, 0.001)
y = 10.0 * np.sin(x) + 1.234
```

You can intersperse text and images with the Python code. In this case, I'm importing the needed libraries and then generating arrays x and y, where y is given by the function:

$$
\begin{eqnarray}
y \left(x\right) & = & 10.0 \sin (x) + 1.234
\end{eqnarray}
$$

```python
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title ("Such Graph. Wow.")
```
